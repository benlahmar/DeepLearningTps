```python
from sklearn.datasets import make_moons
from matplotlib import pyplot
from pandas import DataFrame
```

#generer 2 dataset classification



```python
X,y = make_moons(n_samples=100, noise=0.2, random_state=1)
```

```python
df= DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
```

```python
df.head()
```

```python
colors={0:'red',1:'blue'}
fig, ax = pyplot.subplots()
gr= df.groupby('label')
for key, g in gr:
    g.plot(ax=ax, kind='scatter',x='x',y='y', label=key, color=colors[key])
pyplot.show()
```

```python
n_train= 30
trainX, testX = X[:n_train, :] , X[n_train:, :]
trainy, testy = y[:n_train] , y[n_train:]
```

definir le model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model= Sequential()
model.add(Dense(500, input_dim=2, activation='relu'))
model.add(Dense(1,  activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
 history = model.fit(trainX,trainy, validation_data=(testX, testy), epochs=3000,verbose=1)
```

```python
xx, train_acc = model.evaluate(trainX, trainy, verbose=0)
yy, test_acc = model.evaluate(testX, testy, verbose=0)
print('train:  %.3f, Test : %.3f' % (train_acc, test_acc))
    
```

```python
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```

```python
model.summary()
```

```python
from keras.regularizers import l2
modelrg= Sequential()
modelrg.add(Dense(500, input_dim=2, activation='relu',kernel_regularizer=l2(0.001)))
modelrg.add(Dense(1,  activation='sigmoid'))
```

```python
modelrg.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
modelrg.summary()
```

```python
history = modelrg.fit(trainX,trainy, validation_data=(testX, testy), epochs=3000,verbose=0)
```

```python
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
pyplot.show()
```

```python
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```
